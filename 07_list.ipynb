 ## Introduction of List 

A list in Python is a built-in data type used to store multiple items in a single variable. Lists are ordered, mutable (changeable), and allow duplicate values. Elements in a list are enclosed within square brackets [] and separated by commas. 

## Fearture of list 

- ordered
- mutable
-  allow duplicate
-  heterogenenous
-  indexed
-  dynamic size 

## Advantage of list 

- Lists consume more memory compared to tuples.
- Searching in a large list can be slower.
- Inserting or deleting elements in the middle of a large list takes more time.
- Lists are mutable, so data can be changed accidentally.

## List Methods  

- Method   	  Description
- 1.append()	  Add an item
- 2.insert()	  Insert at position
- 3.extend()	  Add multiple items
- 4.remove()	  Remove specific item
- 5. pop()	    Remove by index
- 6. clear()	    Remove all items

```python
numbers=[1,2,3,4,5,6]
even=[]
odd=[]

for i in numbers:
    if i%2==0:
        even.append(i)
    else:
        odd.append(i)

        print("even:",even)
        print("odd:",odd)
```
