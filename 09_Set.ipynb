## Introduction of Set 

A Set in Python is a collection of unique and unordered elements. It does not allow duplicate values.

## Advantage of set

- Stores only unique values automatically

-  Fast searching (in operator is very efficient)

-  Supports mathematical operations such as union, intersection, and difference

-  Useful for removing duplicates from lists 

## Disadvantage of set 

- Elements are unordered

-  No indexing or slicing

-  Only immutable objects can be stored (lists cannot be elements) 


## Set Methods  

- Add
- Remove
- Update
- Pop
- Clear
- Copy 


## Examples:- 

```python
fruits = {"apple", "banana", "mango"}
print(fruits)
```

```python
numbers1 = {1, 2, 2, 3, 4, 4, 5}

print(numbers1)
numbers2 = {1, 6, 6, 3, 4, 4, 5}

print(numbers2)
```

```python
colors = {"red", "green"}

colors.add("blue")

print(colors)
```

```python
colors = {"red", "green"}

colors.update(["blue", "yellow"])

print(colors)
```

```python
A = {1, 2, 3}
B = {3, 4, 5}

result = A.union(B)

print(result) 
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A.intersection(B))
```

```python
class_a = {"Raj", "Amit", "Rohan"}
class_b = {"Rohan", "Amit", "Sohan"}

common = class_a.intersection(class_b)

print(common)
```

```python
names = ["Raj", "Amit", "Raj", "Rohan", "Amit"]

unique_names = set(names)

print(unique_names)
```

```python
sentence = "python is easy and python is powerful"

words = sentence.split()

unique_words = set(words)

print(unique_words)
print("Total Unique Words:", len(unique_words))
```
