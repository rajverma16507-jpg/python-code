## Introduction of Strig

A string is a sequence of characters enclosed in single quotes, double quotes, or triple quotes.

 ## Feature of String 

- .Immutable (cannot be changed)
- .Ordered
- .Supports indexing
- .Supports slicing
- .Can store letters, digits, symbols, and spaces

 ## Advantage of String 

- .Easy to handle text data
- .Many built-in methods are available
- .Useful for storing names, messages, and sentences.

 ## Disadvantage of String 

- .Immutable in nature
- .Large string operations may consume more memory 

## Type of String:-

- .Single-Quoted String (' ')
- .Double-Quoted String (" ")
- .Triple-Quoted String (''' ''' or """ """) 

 ## Indexing in String 

indexing is the process of accessing individual characters in a string by using their position numbers called indexes.

 ## Example:-

```python
text = "Python"

print(text[0])   # P
print(text[1])   # y
print(text[2])   # t
print(text[-1])  # n
print(text[-2])  # o
```

## Slicing in String 

Slicing is the process of extracting a portion (substring) of a string using a range of indexes. 

 ## Example:-

```python
text = "Python"

print(text[0:4]) 
```
