## Introduction of tuple

A tuple is an ordered collection of elements in Python. It can store different types of data, such as integers, strings, and floats. Unlike lists, tuples are immutable, which means their values cannot be changed after creation. 

 ## Feature of tuple

- .Ordered  
- .Immutable .
- .Allows Duplicates 
- .Supports Indexing 
- .Supports Slicing 

 ## Advantage of tuple

- .Faster than lists for storing fixed data.
- .Uses less memory compared to lists.
- .Data remains safe because tuples cannot be modified.
- .Can be used as dictionary keys. 

 ## Disadvantage of tuple

- .Elements cannot be added, removed, or changed.
- .Less flexible than lists.
- .To modify a tuple, a new tuple must be created. 

 ## Tuple methods

- . tup.index(el)
- .tup.count(el)

## Examples:-

```python
number=(10,20,30,40,50)
length=len(number)
print("length of tuples:",length)
```

```python
number=(1,2,3,4,56,7,8,7,5,7,6,7,6,7,9)
count=number.count(7)
print("number of occurrence :",count)
```

```python
fruits=("apple","bananana","mango","orange")
position=fruits.index("mango")
print("index of mango:",position)
```

```python
t = (15, 45, 10, 80, 25)
print("Maximum:", max(t))
```

```python
t = (1, 2, 3, 4, 5)

if 3 in t:
    print("Element Found")
else:
    print("Element Not Found")
```

```python
my_list = [10, 20, 30, 40]
my_tuple = tuple(my_list)
print(my_tuple)
```

```python
t = (10, 20, 30, 40, 50)
reversed_tuple = t[::-1]
print(reversed_tuple)
```
