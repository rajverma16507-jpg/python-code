 ## Introduction of pyhton:-
Python is a high-level, interpreted, and general-purpose programming language known for its simplicity and readability. It was created by Guido van Rossum and first released in 1991.

 ## Features of Python :-

1. Simple and Easy to Learn
2.  Interpreted Language 
3. High-Level Language 
4. Object-Oriented Programming (OOP) 
5. Platform Independent 
6. Dynamically Typed 

 ## Advantages of Python :-
1. Easy to learn and use
2.Large library ecosystem
3.Portable across platforms
4.Supports multiple programming paradigms
5. Strong community support

 ## Disadvantages of Python :-

1. Slower than compiled languages like C++
2. Higher memory consumption
3. Not ideal for mobile app development
