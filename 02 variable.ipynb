## Introduction of variable 

A variable is a neme used to store data in memory. the value  of a variable can be changed during the execution

# Syntax :- 

variable_name=value

 ## Rule of naming variable 

1. Variable name must start with a letter or underscore(_).
2. It cannot start with a number.
3. Special character like @,#,% are not allowed.
4. Variable are case sensitive. 

 ## Type of variable 

- Integer
- Float
- String
- Boolean(bool)
- Complex

## Advantage of variable  

1.Easy to understand
2.Code reusability
3.Easy modification
4.Impoves readability

## Disadvantage of variable

1.Memory usage
2.Naming confusion
3.Runtime error
4.Dynamic typing issues

```python
name="raj verma" # string
age=20           #integer
height=5.7       #Float
marks= 85+5j     #complex

print(name)
print(age)
print(height)
print(marks)
```
