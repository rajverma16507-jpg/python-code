 ## Introduction of keywards

 Keywords are reserved words in Python that have a special meaning and cannot be used as variable names, function names, or identifiers.

 ## Advantage of keywards 

- .Improve readbility
- .Provide predefined functionality
- .maintain consistency 

 ## Disadvantage of keywards 

 - .Cannot be used as identifier.
- .Need to memorize them. 

 ## Example of keywards  

- .If
- . Else
- .for
- .while
- .Def
- .Return
- .Class
- .True
- .False
- .None
- .Import
- .Try
